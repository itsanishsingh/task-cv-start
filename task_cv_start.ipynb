```python
def draw_circle(event, x, y, flags, params):
    if event == 1:
        cv2.circle(blank_img, center=(x, y), radius=50, thickness=2, color=(0, 0, 255))

cv2.namedWindow(winname="window")
cv2.setMouseCallback("window", draw_circle)

blank_img = np.zeros((512, 512, 3))

while True:
    cv2.imshow("window", blank_img)

    if cv2.waitKey(1) & 0xFF == ord("x"):
        break

cv2.destroyAllWindows()
```

```python
flag = False
ix = -1
iy = -1


def draw_rectangle(event, x, y, flags, params):
    global flag, ix, iy
    if event == 1:
        flag = True
        ix = x
        iy = y

    elif event == 0:
        if flag:
            cv2.rectangle(
                blank_img, pt1=(ix, iy), pt2=(x, y), color=(0, 255, 255), thickness=-1
            )

    elif event == 4:
        flag = False
        cv2.rectangle(
            blank_img, pt1=(ix, iy), pt2=(x, y), color=(0, 255, 255), thickness=-1
        )


cv2.namedWindow(winname="window")
cv2.setMouseCallback("window", draw_rectangle)

blank_img = np.zeros((512, 512, 3))

while True:
    cv2.imshow("window", blank_img)

    if cv2.waitKey(1) & 0xFF == ord("x"):
        break

cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*"MJPG")
out = cv2.VideoWriter("data/output.avi", fourcc, 20.0, (640, 480))

while True:
    ret, frame = cap.read()
    if ret:
        out.write(frame)
        cv2.imshow("window", frame)

        if cv2.waitKey(1) & 0xFF == ord("x"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
